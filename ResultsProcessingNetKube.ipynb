Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('FinalResultsMaxOutYOURLS.csv', index_col=0, header=0)
df = pd.DataFrame(data)
print(df)
```

Plot Average Shorten Times Together

```python
def plotMetric(metric):
  metric = metric
  row = df.loc[metric]

  # plot
  plt.figure(figsize=(10,6))
  colors = plt.cm.tab10(range(len(row)))

  # Scatter plot with legend
  for i, (col_name, value) in enumerate(row.items()):
    plt.scatter(i, value, color=colors[i], label=col_name, s=100)

  # Plot settings
  plt.title(f'{metric} Across Deployment Types')
  plt.ylabel(metric)
  plt.xticks([])  # Hide x-ticks since labels are in the legend
  plt.grid(True)
  plt.legend(title="Configuration", bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.tight_layout()
  plt.show()
```

```python
for metric in df.index:
  plotMetric(metric)
```
